```python
# Import the scraper class
import sys
import pandas as pd
import os
sys.path.insert(0, os.path.abspath('..'))  # add the parent folder
from scraper import LinkedInScraper
```

```python
# LinkedIn search parameters
keyword = "Data Analyst"
location = "Addis Ababa"
chromedriver_path = r"C:\Windows\chromedriver.exe"
max_jobs = 20  # set how many jobs to scrape
```

```python
from urllib.parse import quote_plus

q = quote_plus(keyword)
loc = quote_plus(location)

url = f"https://www.linkedin.com/jobs/search/?keywords={q}&location={loc}"
```

```python
scraper = LinkedInScraper(url=url, chromedriver_path=chromedriver_path, max_jobs=max_jobs)
```

```python
# Start Chrome and open LinkedIn
scraper.setup_driver(headless=False)  # headless=False so you can login manually
scraper.open_page()
scraper.scroll_page(times=5)          # scroll a few times to load jobs
```

```python
scraper.extract_jobs()                # scrape jobs
df = scraper.get_jobs_dataframe()     # get results as DataFrame
scraper.close_driver()
```
